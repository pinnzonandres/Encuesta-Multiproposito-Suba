## NVCBP6
### 6. ¿Cuántos pisos tiene la edificación donde está ubicada la vivienda?

**Datos: 24536**

```python
data['NVCBP6'].describe()
```

```python
# creating a figure composed of two matplotlib.Axes objects (ax_box and ax_hist)
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# assigning a graph to each ax
sns.boxplot(x = data['NVCBP6'], ax=ax_box)
#ax_box.boxplot(data['NVCBP6'], vert = False)
#sns.histplot(data=data, x='NVCBP6', ax=ax_hist)
ax_hist.hist(data['NVCBP6'], bins = np.arange(32) - 0.5, edgecolor = 'k')
ax_hist.axvline(data['NVCBP6'].mean(), color='r', linestyle='dashed', linewidth=1, label = 'Media')
ax_hist.axvline(data['NVCBP6'].median(), color='b', linestyle='dashed', linewidth=2, label = 'Mediana')
ax_hist.axvline(data['NVCBP6'].quantile(0.25), color='k', linestyle=':', linewidth=1)
ax_hist.axvline(data['NVCBP6'].quantile(0.75), color='k', linestyle=':', linewidth=1)
ax_hist.legend()
# Remove x axis name for the boxplot
ax_box.set_title('¿Cuántos pisos tiene la edificación donde está ubicada la vivienda?')
ax_hist.set_xticks(range(0,32,2), align = 'center')
ax_box.set(xlabel='')
plt.show()
```
