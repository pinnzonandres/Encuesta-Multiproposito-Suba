```python
primos = [2,3,5,7]
for i,j in zip(primos, comercio[1:]):
    print(i,j)
```

```python
def find_prime_numbers(n:int):
    list_of_prime=[]
    if n == 1:
        list_of_prime.append(1)
        return list_of_prime
    else:
        i = 2
        while n!=1:
            while n%i == 0:
                n = n / i
                if i not in list_of_prime:
                    list_of_prime.append(i)        
            i = i + 1
        return list_of_prime
```

```python
find_prime_numbers(18)
```

```python
dict_comercio = {2:"Comercio",
                3:"Industria",
                5:"Servicios",
                7:"Agropecuario",
                1:"No está dedicado a ningún negocio"}
```

```python
def get_comercio(row):
    a = 1
    for i,j in zip(primos, comercio[1:]):
        if row[j] == 0 or row[j] == 2:
            a *= 1
        else:
            a *= i*row[j]
    list_ = find_prime_numbers(a)
    list_ = [dict_comercio[i] for i in list_]
    val = " - ".join(list_)
    return val
```

```python
data['Tipo de Comercio'] = data.apply(get_comercio, axis = 1)
```
