### NPCHP18
En un trayecto normal a su sitio de estudio, ¿cuáles medios de transporte utiliza …:

**Contestan 5506**

```python
transporte = [i for i in data.columns if 'NPCHP18' in str(i)]
print(transporte)
```

```python
for i in transporte:
    print(i, data[i].count())
```

```python
for i in transporte:
    data[i] = data.apply(lambda x: 0 if (x[i] == 2) else (1 if (x[i] == 1) else 2), axis = 1) 
```

```python
list_tra = ['Transmilenio',
            'Buses del Sitp',
            'Bus buseta o colectivo',
            'Automovil de uso particular',
            'Taxi',
            'Motocicleta',
            'Bicicleta',
            'Ruta Escolar',
            'A pie',
            'Bus Intermunicipal',
            'Bicitaxi o mototaxi',
            'Aplicaciones',
           'Vehículo, patineta, bicicleta o  moto eléctricos',
           'Caballo',
           'Otro','No se desplaza']

colors = ['#264653','#e63946','#ddbea9','#0077b6','#283618','#fca311','#ffc6ff','#219ebc','#d62828','#3d405b','#7209b7','#d4a373','#fb5607','#2C407B','#FFD400','#F58220']
```

```python
for i,j,z in zip(transporte,list_tra,colors):
    plt.figure(figsize=(10,5))
    bars = plt.bar(data[i].value_counts().index.tolist(),data[i].value_counts().tolist(), edgecolor = 'black', color = z)
    plt.title(f'Pregunta {i} \n En un trayecto normal a su sitio de estudio, ¿cuáles medios de transporte utiliza …: \n {j}?')
    plt.xticks([0,1,2],['No','Si', 'No Contesta'])
    plt.bar_label(bars)
    plt.show()
```
