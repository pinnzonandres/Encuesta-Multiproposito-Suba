### NVCBP6
¿Cuántos pisos tiene la edificación donde está ubicada la vivienda?

**Todos los datos**

```python
data['NVCBP6'].describe()
```

```python
# creating a figure composed of two matplotlib.Axes objects (ax_box and ax_hist)
f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw={"height_ratios": (.15, .85)})
 
# assigning a graph to each ax
sns.boxplot(x = data['NVCBP6'], ax=ax_box)
#ax_box.boxplot(data['NVCBP6'], vert = False)
#sns.histplot(data=data, x='NVCBP6', ax=ax_hist)
ax_hist.hist(data['NVCBP6'], bins = np.arange(32) - 0.5, edgecolor = 'k')
ax_hist.axvline(data['NVCBP6'].mean(), color='r', linestyle='dashed', linewidth=1, label = 'Media')
ax_hist.axvline(data['NVCBP6'].median(), color='b', linestyle='dashed', linewidth=2, label = 'Mediana')
ax_hist.axvline(data['NVCBP6'].quantile(0.25), color='k', linestyle=':', linewidth=1)
ax_hist.axvline(data['NVCBP6'].quantile(0.75), color='k', linestyle=':', linewidth=1)
ax_hist.legend()
# Remove x axis name for the boxplot
ax_box.set_title('¿Cuántos pisos tiene la edificación donde está ubicada la vivienda?')
ax_hist.set_xticks(range(0,32,2), align = 'center')
ax_box.set(xlabel='')
plt.show()
```

### NVCBP14
La vivienda está cerca de:

- a. Fábricas o Industrias
- b. Basureros o botaderos de basuras
- c. Plazas de Mercado o Mataderos
- d. Terminales de Buses
- e. Bares o discotecas
- f. Expendios de droga (ollas)
- g. Lotes baldíos o sitios oscuros y peligrosos
- h. Lineas de alta tensión
- i. Caños de aguas resiguales
- j. Zona de riesgo de incendio forestal
- k. Talleres de mecánica, servitecas o estaciones de gasolina
- l. Prostíbulos

**Todas las preguntas fueron contestadas en su totalidad**

```python
problemas = [i for i in data.columns if ('NVCBP14') in i]
```

```python
for i in problemas:
    print(i, data[i].count())
```

```python
for i in problemas:
    data = data.replace({i:2},0)
```

```python
list_pro = ['Fábricas o Industrias',
            'Basureros o botaderos de basuras',
            'Plazas de Mercado o Mataderos',
            'Terminales de Buses',
            'Bares o discotecas',
            'Prostíbulos',
            'Expendios de droga (ollas)',
            'Lotes baldíos o sitios oscuros y peligrosos',
            'Lineas de alta tensión',
            'Caños de aguas resiguales',
            'Zona de riesgo de incendio forestal',
            'Talleres de mecánica, servitecas o estaciones de gasolina']

colors = ['#264653','#e63946','#ddbea9','#0077b6','#283618','#fca311','#ffc6ff','#219ebc','#d62828','#3d405b','#7209b7','#d4a373','#fb5607']
```

```python
fig, axes = plt.subplots(1,3, figsize = (15,5))
fig.subplots_adjust(top=0.9)
fig.suptitle('La vivienda está cerca de:')
for ax,cols,names in zip(axes,problemas,list_pro):
    sns.countplot(x = cols, data = data, ax = ax)
    ax.grid(alpha = 0.2, axis = 'y')
    ax.set_title(f'{names}')
    ax.set_xticks([0,1],['No','Si'])
    ax.margins(y=0.1) # make room for the labels
    for bars in ax.containers:
        ax.bar_label(bars, fmt='%.0f', fontsize=9)
plt.tight_layout()
    
fig, axes = plt.subplots(1,3, figsize = (15,5))
for ax,cols,names in zip(axes,problemas[3:],list_pro[3:]):
    sns.countplot(x = cols, data = data, ax = ax)
    ax.grid(alpha = 0.2, axis = 'y')
    ax.set_title(f'{names}')
    ax.set_xticks([0,1],['No','Si'])
    ax.margins(y=0.1) # make room for the labels
    for bars in ax.containers:
        ax.bar_label(bars, fmt='%.0f', fontsize=9)
plt.tight_layout()

fig, axes = plt.subplots(1,3, figsize = (15,5))
for ax,cols,names in zip(axes,problemas[6:],list_pro[6:]):
    sns.countplot(x = cols, data = data, ax = ax)
    ax.grid(alpha = 0.2, axis = 'y')
    ax.set_title(f'{names}')
    ax.set_xticks([0,1],['No','Si'])
    ax.margins(y=0.1) # make room for the labels
    for bars in ax.containers:
        ax.bar_label(bars, fmt='%.0f', fontsize=9)
plt.tight_layout()
    
fig, axes = plt.subplots(1,3, figsize = (15,5))
for ax,cols,names in zip(axes,problemas[9:],list_pro[9:]):
    sns.countplot(x = cols, data = data, ax = ax)
    ax.grid(alpha = 0.2, axis = 'y')
    ax.set_title(f'{names}')
    ax.set_xticks([0,1],['No','Si'])
    ax.margins(y=0.1) # make room for the labels
    for bars in ax.containers:
        ax.bar_label(bars, fmt='%.0f', fontsize=9)
plt.tight_layout()    
    
plt.show()
```
