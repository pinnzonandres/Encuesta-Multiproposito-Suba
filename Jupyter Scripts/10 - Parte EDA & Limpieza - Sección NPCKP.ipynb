# Novena Sección (NPCKP)

## NPCKP1
### 1. ¿En qué actividad ocupó ...la mayor parte del tiempo la SEMANA PASADA?

1. Trabajando
2. Buscando trabajo
3. Estudiando
4. Oficios del hogar
5. Incapacitado(a) permanente para trabajar
6. Otra actividad

**Contestan 22213**

```python
data['NPCKP1'].count()
```

```python
data['NPCKP1'].value_counts()
```

```python
data['NPCKP1'].describe().apply("{0:.2f}".format)
```

```python
fig, ax = plt.subplots(figsize=(18, 5))

g = sns.countplot(ax=ax, data = data, x ='NPCKP1')

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f', fontsize=9)
            
ax.set_title('¿En qué actividad ocupó ...la mayor parte del tiempo la SEMANA PASADA?')
ax.set_xticklabels(['Trabajando', 'Buscando trabajo', 'Estudiando','Oficios del hogar', 
                    'Incapacitado(a) permanente\n para trabajar', 'Otra actividad'])

plt.grid(alpha = 0.2, axis = 'y')
plt.show()
```
